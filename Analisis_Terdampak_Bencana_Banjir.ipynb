# **Analisis Statistik Deskriptif dan Sebaran Temporal Desa/Kelurahan Terdampak Bencana Banjir di Provinsi Jawa Timur Tahun 2019-2021**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat data ke dalam DataFrame
data = {
    'Regency_City': ['Pacitan', 'Ponorogo', 'Trenggalek', 'Tulungagung', 'Blitar', 'Kediri', 'Malang',
                     'Lumajang', 'Jember', 'Banyuwangi', 'Bondowoso', 'Situbondo', 'Probolinggo',
                     'Pasuruan', 'Sidoarjo', 'Mojokerto', 'Jombang', 'Nganjuk', 'Madiun', 'Magetan',
                     'Ngawi', 'Bojonegoro', 'Tuban', 'Lamongan', 'Gresik', 'Bangkalan', 'Sampang',
                     'Pamekasan', 'Sumenep', 'Kediri (Kota)', 'Malang (Kota)', 'Pasuruan (Kota)', 'Surabaya'],
    '2019': [18, 49, 26, 3, 6, 14, 8, 11, 30, 18, 7, 14, 6, 73, 53, 30, 67, 25, 22, 4, 20, 75, 45, 65, 48, 5, 10, 10, 5, 4, 4, 18, 7],
    '2020': [13, 30, 26, 9, 6, 4, 2, 16, 28, 7, 8, 2, 14, 72, 24, 20, 67, 42, 39, 1, 14, 47, 29, 79, 53, 9, 5, 12, 6, 1, 6, 21, 1],
    '2021': [16, 49, 15, 18, 9, 42, 13, 14, 39, 23, 8, 3, 19, 83, 57, 29, 73, 62, 45, 4, 23, 52, 52, 104, 62, 5, 14, 19, 11, 3, 7, 18, 1]
}
df = pd.DataFrame(data)

# Mengatur tema visual secara global
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))

# Membuat KDE plot untuk masing-class tahun
sns.kdeplot(data=df['2019'], fill=True, label='2019', alpha=0.5)
sns.kdeplot(data=df['2020'], fill=True, label='2020', alpha=0.5)
sns.kdeplot(data=df['2021'], fill=True, label='2021', alpha=0.5)

# Pengaturan label dan judul
plt.title('Distribution of Village Floods across East Java Regencies (2019-2021)')
plt.xlabel('Number of Villages Affected')
plt.ylabel('Density')
plt.legend()

plt.show()
```

```python
# Mengurutkan data dan mengambil 10 data teratas
top10_2021 = df[['Regency_City', '2021']].sort_values(by='2021', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Membuat bar chart horizontal
sns.barplot(x='2021', y='Regency_City', data=top10_2021, palette='coolwarm', hue='Regency_City', legend=False)

# Pengaturan label dan judul
plt.title('Top 10 Most Affected Areas in 2021')
plt.xlabel('Villages Affected')
plt.ylabel('Regency/City')

plt.show()
```

```python
# Menghitung total untuk masing-masing tahun
totals = df[['2019', '2020', '2021']].sum()

plt.figure(figsize=(8, 6))

# Membuat bar chart dengan warna spesifik
ax = sns.barplot(x=totals.index, y=totals.values, palette=['#3498db', '#95a5a6', '#e74c3c'], hue=totals.index, legend=False)

# Menambahkan judul dan label
plt.title('Total Number of Villages Impacted by Floods (2019-2021)')
plt.ylabel('Total Affected Villages')

# Menambahkan teks angka di atas bar
for i, v in enumerate(totals.values):
    ax.text(i, v + 10, str(v), ha='center', va='bottom', fontweight='bold', fontsize=12)

# Mengatur batas atas sumbu Y agar label teks tidak terpotong
plt.ylim(0, 1100)

plt.show()
```
